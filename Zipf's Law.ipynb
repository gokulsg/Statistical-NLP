# Zipf's Law

### Zipf's law is a relation between rank order and frequency of occurrence. According to Zipf's law, the frequency of a given word is inversely proportional to its rank. This law describes how tokens are distributed in languages like some occur more frequently where as some occur very rarely. 

### The python library nltk is used for getting the frequency distribution and Matplotlib for plotting the charts. Here, the words on the corpus are tokenized by considering the whitespace using .split(). Also the charts were plotted using linear and log scale.

```python
import nltk
import matplotlib.pyplot as plt

# Read the input file and output the list of tokens #
def generate_tokens(file_name):
    """
    Generate tokens by processing the input file

    Args:
        file_name (string): file to be processed
        
    Returns:
        words(list): list of word obtained after word tokenization
    """
    file = open(file_name, 'r', encoding='utf-8')
    words = [x.lower() for x in file.read().split()]
    return words

# Get the frequency distribution and sort the dictionary in decending order of frequency and output the frequency list#
def desc_sort_freq(words):
    """
    Get the frequency distribution and sort the dictionary in decending order of frequency and output the frequency list

    Args:
        words (list): list of word obtained after word tokenization
        
    Returns:
        (list): list of words sorted in decending order of frequencies
    """
    freq_dict = nltk.FreqDist(words)
    # Sorting dictionary in descending order of frequency #
    # Ref : https://stackoverflow.com/questions/613183/how-do-i-sort-a-dictionary-by-value #
    freq_dict = dict(sorted(freq_dict.items(), key=lambda item: item[1], reverse=True))
    # returns list of frequencies sorted in descending order #
    return list(freq_dict.values())

# Ploting the Graph #
def plot_graph(freq_lst, corpus_name = 'corpus', num_frequency = None, plot = 'linear'):
    """
    Display the result as Graph

    Args:
        freq_lst (list): list of words sorted in decending order of frequencies
        corpus_name (string) : Name of the corpus used
        num_frequency (int) : Number of words to be considered for plotting
        plot (string) : To specify linear or loglog plot. 
    """
    fig, ax = plt.subplots()
    ax.set_xlabel("words")
    
    if num_frequency is not None and num_frequency < len(freq_lst):
        freq_lst = freq_lst[0:num_frequency]
    
    num_words = range(1, len(freq_lst)+1)
        
    if plot == 'loglog':
        ax.loglog(num_words, freq_lst, label=corpus_name)
        ax.set_ylabel("log(Frequency)")
        ax.set_title("Frequency Curve with log scaling for the corpus "+corpus_name)
        
    elif plot == 'linear':
        ax.plot(num_words, freq_lst, label=corpus_name)
        ax.set_ylabel("Frequency")
        ax.set_title("Frequency Curve with linear scaling for the corpus "+corpus_name)
    
    ax.grid()
    plt.show() 
```
